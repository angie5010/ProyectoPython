```python
import pandas as pd
import matplotlib.pyplot as mpl 

# Cargamos los datos limpios.
# Ruta del archivo con los datos limpios.
rutaDatos = "../datos/procesados/productos-limpios.csv" 
# leer datos de csv y guardar en variable
datos = pd.read_csv(rutaDatos) 

# Analisis y visualizacion
print("Estadísticas Básicas") 
# Pequeño resumen estadístico del DataFrame.
print(datos.describe()) 

# Creación de histograma de precios.
#Creamos un histograma de la columna "Price" con 20 barras de color verde y bordes negros
mpl.hist(datos["price"], bins=20, color="green", edgecolor="black") 
# Definición de título para histograma
mpl.title("Distribución de precios") 
# Definición etiqueta para el eje X y Y
mpl.xlabel("Precio") 
mpl.ylabel("Frecuencia") 
# Mostrar el histograma creado
mpl.show() 

# Pequeña visualización de productos más caros
print("Productos más Caros: ") 
productosMasCaros= datos.nlargest(5,'price')
print(productosMasCaros)

```
